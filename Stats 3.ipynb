# Z- Test

```python
import numpy as np
import scipy.stats as stats
```

```python
sample_mean = 110
population_mean = 100
population_std = 15
sample_size = 50
alpha = 0.05
```

```python
z_score = (sample_mean-population_mean)/(population_std/np.sqrt(sample_size))
z_score
```

```python
confidence_interval = 1-0.05
confidence_interval
```

```python
z_critical = stats.norm.ppf(confidence_interval)
z_critical
```

```python
if z_score>z_critical:
    print("We reject the null hypothesis")
else:
    print("We fail to reject the null hypothesis")
```

# T test

1. One sample T test w.r.t. one independent sample
2. Two sample T test w.r.t. two independent samples
3. Paired T test -> Two samples from the same population at different timestamps or time intervals.

### One sample T test

```python
my_cricket_score = [22,38,19,15,48,11,10,49,47,38,10,25,46,10,21,24,29,36,25,25] # this is our sample
len(my_cricket_score)
```

```python
np.mean(my_cricket_score) # sample mean
```

null_hypothesis = sample mean should be equal to population mean = 27.4

```python
import scipy
scipy.stats.ttest_1samp(my_cricket_score,15) # we assume the population mean as 15 
# now 15 is very far from the sample mean which is 27.4
```

```python
p_value = scipy.stats.ttest_1samp(my_cricket_score,15)[1]
p_value
```

```python
alpha = 0.05
```

```python
if p_value<alpha:
    print("We reject the null hypothesis.")
else:
    print("We fail to reject the null hypothesis.")
```

```python
# if say population mean was 26
p_value = scipy.stats.ttest_1samp(my_cricket_score,26)[1]
p_value
# this is very close to the sample mean 
```

```python
if p_value<alpha:
    print("We reject the null hypothesis.")
else:
    print("We fail to reject the null hypothesis.")
```

### Two sample T test

```python
# two independent samples from different populations
my_cricket_score = [22,38,19,15,48,11,10,49,47,38,10,25,46,10,21,24,29,36,25,25]
virat_cricket_score = [33,45,23,25,46,46,46,49,49,44,44,49,25,31,25,40,30,20,42,37]
```

```python
first_sample_mean = np.mean(my_cricket_score)
second_sample_mean = np.mean(virat_cricket_score)
```

```python
first_sample_mean
```

```python
second_sample_mean
```

null_hypothesis = both sample means would be equal

```python
scipy.stats.ttest_ind(my_cricket_score,virat_cricket_score)
```

```python
p_value = scipy.stats.ttest_ind(my_cricket_score,virat_cricket_score)[1]
p_value
```

```python
alpha = 0.05
if p_value<alpha:
    print("We reject the null hypothesis.")
else:
    print("We fail to reject the null hypothesis.")
# the two means are far from each other
```

### Paired sample T test

```python
import random

population = []
for i in range(1,51):
    population.append(random.randint(10,80))
len(population)
```

```python
my_cricket_score_from_first_innings = np.random.choice(population,20)
```

```python
my_cricket_score_from_second_innings = np.random.choice(population,20)
```

```python
# We take the samples from the same population on different time intervals
my_cricket_score_from_first_innings
```

```python
my_cricket_score_from_second_innings
```

```python
np.mean(my_cricket_score_from_first_innings)
```

```python
np.mean(my_cricket_score_from_second_innings)
```

null hypothesis -> Both the samples will have same mean

```python
scipy.stats.ttest_rel(my_cricket_score_from_first_innings,my_cricket_score_from_second_innings)
```

```python
p_value = scipy.stats.ttest_rel(my_cricket_score_from_first_innings,my_cricket_score_from_second_innings)[1]
p_value
```

```python
alpha = 0.05

if p_value < alpha:
    print("We reject the null hypothesis")
else:
    print("We fail to reject the null hypothesis")
```

#### The weights of 25 obese people were taken before enrolling them in nutrition camp. The population mean weight was found to be 45 kg before starting the camp. After finishing the camp, for the same 25 people, the sample mean was found to be 75 with a standard deviation of 25. Did the fitness camp work ? [CI = 95 %]

```python
population_mean = 45
sample_mean = 75
sample_size = 25
sample_std = 25
```

```python
df = sample_size-1
df
```

1. H0 = population_mean = 45
2. H1 = population_mean =/= 45

```python
t_stats = (sample_mean-population_mean)/(sample_std/np.sqrt(sample_size))
t_stats
```

```python
alpha = 0.05
CI = 0.95
```

```python
critical_value = stats.t.ppf(CI,df)
critical_value
```

```python
print("t_stats = ",t_stats)
print("Critical value = ",critical_value)
```

```python
if t_stats > critical_value:
    print("We reject the null hypothesis")
else:
    print("We fail to reject the null hypothesis")
```

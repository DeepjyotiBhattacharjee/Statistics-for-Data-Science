# QQ Plot

Quantile - Quantile Plot

```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab

def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample,kde=True)
    
    plt.subplot(1,2,2)
    stat.probplot(sample,dist="norm",plot=pylab)
    plt.show()
```

```python
plot_data(log_normal_distributed_data)    
```

```python
plot_data(normal_distributed_data)
```

## Probability mass function ( Discrete Probability Distribution ).
 1. Uniform
 2. Bernoulli
 3. Binomial Distribution.
 4. Poisson distribution.
## Probability density function ( Continuous Probability Distribution ).
 1. Uniform
 2. normal

 4. Log normal
 5. Power Law
 6. Pareto

```python
from numpy import random
random.uniform(size=500) # Continuous values
```

```python
sns.distplot(random.uniform(size=500),hist=False)
```

```python
np.random.randint(1,8,size=800) # Discrete values
```

```python
sns.histplot(np.random.randint(1,8,size=800)) # Uniform distribution
```

```python
p = 0.3
sample_size = 1000
bernoulli_data = np.random.choice([0,1],size=sample_size,p=[1-p,p])

sns.histplot(bernoulli_data,bins=2,kde=False,color="blue")
```

```python
no_of_trials = 10
prob = 0.6
no_of_samples = 10000
binomial_dist = np.random.binomial(no_of_trials,prob,no_of_samples)
sns.distplot(binomial_dist)
```

```python
poisson_data = np.random.poisson(lam = 5,size=1000)

# Plotting a histogram
ax = sns.distplot(poisson_data,
                  kde=False,
                  color="blue")

ax.set(xlabel = "Poisson Distribution",ylabel = "Frequency")
plt.show()
```

# Central Limit Theorem

```python
population = np.random.randint(10,20,50)
population
```

```python
np.mean(population)
```

```python
np.median(population)
```

```python
import statistics
statistics.mode(population)
```

```python
# random sample from population
sample = np.random.choice(population,20)
sample
```

```python
print(f"Mean of the sample = {np.mean(sample)}")
print(f"Median of the sample = {np.median(sample)}")
print(f"Mode of the sample = {statistics.mode(sample)}")
```

### Suppose you have a population with any shape of probability distribution (not necessarily normal), and you draw random samples (at least 30) of a fixed size from this population. As the sample size increases (can be anything, take as much as ou want), the distribution of the sample means approaches a normal distribution, regardless of the shape of the original population distribution. 

```python
sample1 = np.random.choice(population,20)
sample2 = np.random.choice(population,20)
sample3 = np.random.choice(population,20)
sample4 = np.random.choice(population,20)
```

```python
print(sample1)
print(60*"=")
print(sample2)
print(60*"=")
print(sample3)
print(60*"=")
print(sample4)
print(60*"=")
```

```python
mean_of_samples = []
all_samples = [sample1,sample2,sample3,sample4]

for sample in all_samples:
    mean_of_samples.append(np.mean(sample))
mean_of_samples
```

```python
sns.histplot(mean_of_samples,kde=True)
```

```python
population = np.random.binomial(10,0.5,10000)
population
```

```python
len(population)
```

```python
sns.distplot(population)
```

```python
size_of_sample = int(len(population)*0.45) # let's take 45 % of population as sample size
size_of_sample
```

```python
sample_mean = []

for i in range(30):
    sample = np.random.choice(population,size=size_of_sample)
    sample_mean.append(np.mean(sample))
```

```python
sample_mean
```

```python
sns.distplot(sample_mean,kde=True)
```
